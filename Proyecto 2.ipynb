## Introducci√≥n

Zyfra desarrolla soluciones de eficiencia para la industria pesada. El objetivo de este proyecto es construir un modelo que prediga la recuperaci√≥n de oro en dos etapas del proceso de producci√≥n: recuperaci√≥n rougher y recuperaci√≥n final. Esto ayudar√° a optimizar la producci√≥n y eliminar par√°metros no rentables.

- `gold_recovery_train.csv`: conjunto de entrenamiento

# Paso 1.1 Preparar datos.
Abrir y examinar datos.

Vamos a importar las bibliotecas necesarias, cargar los tres archivos y observar su estructura.

```python
# Importaci√≥n de bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci√≥n de estilo
sns.set(style="whitegrid")
pd.set_option('display.max_columns', None)

# Cargar los archivos
train = pd.read_csv('/datasets/gold_recovery_train.csv', index_col='date', parse_dates=True)
test = pd.read_csv('/datasets/gold_recovery_test.csv', index_col='date', parse_dates=True)
full = pd.read_csv('/datasets/gold_recovery_full.csv', index_col='date', parse_dates=True)

# Primer vistazo a cada uno
print("Conjunto de entrenamiento:")
display(train.head())
print("\nConjunto de prueba:")
display(test.head())
print("\nConjunto completo:")
display(full.head())

# Tama√±o de cada conjunto
print("\nTama√±os:")
print("Entrenamiento:", train.shape)
print("Prueba:", test.shape)
print("Completo:", full.shape)
```

# Paso 1.2. Verificar el c√°lculo de recuperaci√≥n (recovery)

Comprueba que el c√°lculo de la recuperaci√≥n sea correcto. Calcula la recuperaci√≥n de la caracter√≠stica (`rougher.output.recovery`) mediante el conjunto de entrenamiento. Encuentra el EAM entre tus c√°lculos y los valores de la caracter√≠stica. Facilita los resultados.

```python
from sklearn.metrics import mean_absolute_error

# Extraer las columnas necesarias
C = train['rougher.output.concentrate_au']
F = train['rougher.input.feed_au']
T = train['rougher.output.tail_au']

# Calcular la recuperaci√≥n seg√∫n la f√≥rmula
recovery_calc = (C * (F - T)) / (F * (C - T)) * 100

# Recuperaci√≥n real
recovery_real = train['rougher.output.recovery']

# Crear un DataFrame para filtrar valores v√°lidos
recovery_df = pd.DataFrame({
    'real': recovery_real,
    'calc': recovery_calc
}).replace([np.inf, -np.inf], np.nan).dropna()

# Calcular MAE
mae = mean_absolute_error(recovery_df['real'], recovery_df['calc'])

print(f'Error Absoluto Medio (MAE) entre recuperaci√≥n calculada y real: {mae:.4f}')
```

```python
from sklearn.metrics import mean_absolute_error

# Extraer las columnas necesarias
C = train['rougher.output.concentrate_au']
F = train['rougher.input.feed_au']
T = train['rougher.output.tail_au']

# Calcular la recuperaci√≥n seg√∫n la f√≥rmula
recovery_calc = (C * (F - T)) / (F * (C - T)) * 100

# Recuperaci√≥n real
recovery_real = train['rougher.output.recovery']

# Crear un DataFrame para filtrar valores v√°lidos
recovery_df = pd.DataFrame({
    'real': recovery_real,
    'calc': recovery_calc
}).replace([np.inf, -np.inf], np.nan).dropna()

# Calcular MAE
mae = mean_absolute_error(recovery_df['real'], recovery_df['calc'])

print(f'Error Absoluto Medio (MAE) entre recuperaci√≥n calculada y real: {mae:.4f}')
```

### 1.3. An√°lisis de caracter√≠sticas no disponibles en el conjunto de prueba

Vamos a comparar las columnas disponibles en el conjunto de entrenamiento (`train`) y el conjunto de prueba (`test`) para identificar qu√© caracter√≠sticas no est√°n presentes en este √∫ltimo.

Esto es importante para asegurarnos de que el modelo no utilice variables que no estar√°n disponibles en la predicci√≥n final.

```python
# Verificar columnas ausentes en el conjunto de prueba
train_cols = set(train.columns)
test_cols = set(test.columns)

# Diferencia: columnas que est√°n en train pero no en test
missing_in_test = train_cols - test_cols

# Mostrar resultados ordenados
missing_in_test = sorted(missing_in_test)
print(f'Columnas no disponibles en el conjunto de prueba ({len(missing_in_test)}):')
for col in missing_in_test:
    print(col)

# Tipos de datos de las columnas ausentes
train[missing_in_test].dtypes
```

#### Resultado:

Se identificaron X columnas que est√°n presentes en el conjunto de entrenamiento pero ausentes en el de prueba. Estas columnas no deben utilizarse para el entrenamiento del modelo.

Estas caracter√≠sticas incluyen objetivos como `rougher.output.recovery` y variables intermedias calculadas despu√©s del experimento.

### 1.4. Preprocesamiento de datos

En este paso realizaremos el preprocesamiento necesario para asegurar la calidad del conjunto de datos antes de construir modelos. Las tareas incluyen:

- Verificamos que los archivos de entrenamiento, prueba y completos contienen la columna `date`.
- Si la columna est√° presente, la convertimos a formato datetime y la usamos como √≠ndice temporal.
- Luego, igualamos las columnas del conjunto de entrenamiento con las del conjunto de prueba, eliminando las columnas no presentes en este √∫ltimo.
- Finalmente, eliminamos cualquier fila con valores faltantes.

```python
# Verificamos las columnas presentes en los archivos
print("Columnas en train:", train.columns[:5])
print("Columnas en test:", test.columns[:5])
print("Columnas en full:", full.columns[:5])

# Verificamos si existe la columna 'date'
for name, df in zip(['train', 'test', 'full'], [train, test, full]):
    if 'date' not in df.columns:
        print(f"La columna 'date' no est√° en {name}")

# Recargamos los datos con seguridad (por si ya ten√≠an 'date' como √≠ndice)
train_data = train.reset_index()
test_data = test.reset_index()
full_data = full.reset_index()

# Convertimos 'date' a datetime y la asignamos como √≠ndice
for df in [train_data, test_data, full_data]:
    df['date'] = pd.to_datetime(df['date'])
    df.set_index('date', inplace=True)

# Igualamos las columnas de entrenamiento con las de prueba (test no tiene columnas objetivo)
train_data = train_data[test_data.columns]  # solo las columnas que est√°n en test

# Revisamos valores nulos antes de eliminar
print("Valores nulos en train:")
print(train_data.isnull().sum().value_counts())

print("\nValores nulos en test:")
print(test_data.isnull().sum().value_counts())

# Eliminamos filas con valores faltantes
train_data = train_data.dropna()
test_data = test_data.dropna()
```

#### Resultado del preprocesamiento (corregido):

- La columna `date` fue correctamente localizada y usada como √≠ndice en todos los DataFrames.
- Se igualaron las columnas del conjunto de entrenamiento a las del conjunto de prueba, excluyendo las no disponibles en este √∫ltimo.
- Se eliminaron filas con valores nulos en ambos conjuntos.
- Ambos DataFrames (`train_data` y `test_data`) est√°n limpios, con columnas alineadas y listos para el an√°lisis exploratorio.

## 2. An√°lisis de los datos

### 2.1 ¬øC√≥mo cambia la concentraci√≥n de metales (Au, Ag, Pb) seg√∫n la etapa de purificaci√≥n?

Vamos a analizar c√≥mo cambia la concentraci√≥n de los metales oro (Au), plata (Ag) y plomo (Pb) a lo largo del proceso de purificaci√≥n. Para ello, trazaremos gr√°ficos de l√≠nea para cada metal en las diferentes etapas:

- `rougher.output.concentrate_*`: primera etapa de flotaci√≥n (rougher).
- `primary_cleaner.output.concentrate_*`: primera etapa de limpieza (primary).
- `final.output.concentrate_*`: salida final del concentrado.

Este an√°lisis ayudar√° a entender el comportamiento de los metales durante el proceso de refinamiento.

```python
import matplotlib.pyplot as plt

# Lista de metales a analizar
metales = ['au', 'ag', 'pb']

# Crear los gr√°ficos
for metal in metales:
    plt.figure(figsize=(12, 4))
    
    # Extraer series por etapa
    rougher = full_data[f'rougher.output.concentrate_{metal}']
    primary = full_data[f'primary_cleaner.output.concentrate_{metal}']
    final = full_data[f'final.output.concentrate_{metal}']
    
    # Graficar
    plt.plot(rougher.index, rougher, label='Rougher')
    plt.plot(primary.index, primary, label='Primary Cleaner')
    plt.plot(final.index, final, label='Final')
    plt.title(f'Concentraci√≥n de {metal.upper()} por etapa de purificaci√≥n')
    plt.xlabel('Fecha')
    plt.ylabel('Concentraci√≥n')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

### 2.2 Comparar las distribuciones del tama√±o de las part√≠culas de alimentaci√≥n en los conjuntos de entrenamiento y prueba

El tama√±o de las part√≠culas en la alimentaci√≥n puede afectar la eficiencia del proceso de extracci√≥n. Vamos a comparar la distribuci√≥n de esta caracter√≠stica en los conjuntos `train` y `test`.

Si las distribuciones son significativamente diferentes, esto puede influir negativamente en la evaluaci√≥n del modelo, ya que estar√≠a evaluando con datos que no representan bien la distribuci√≥n usada en el entrenamiento.

```python
import seaborn as sns

# Variable a comparar
columna_particulas = 'rougher.input.feed_size'

# Gr√°fico de distribuci√≥n
plt.figure(figsize=(10, 5))
sns.kdeplot(train_data[columna_particulas], label='Train', fill=True)
sns.kdeplot(test_data[columna_particulas], label='Test', fill=True)
plt.title('Distribuci√≥n del tama√±o de part√≠culas en la alimentaci√≥n (feed_size)')
plt.xlabel('Tama√±o de part√≠cula')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 2.3 Revisar concentraciones totales de sustancias y detectar valores an√≥malos

Calculamos la suma de las concentraciones de los metales y s√≥lidos en la materia prima.  
Luego, revisamos la estad√≠stica descriptiva para detectar valores fuera de rango (por ejemplo, suma 0 o mayor a 100%).  
Finalmente, visualizamos la distribuci√≥n y eliminamos las filas an√≥malas para limpiar el conjunto de entrenamiento.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular la concentraci√≥n total en la materia prima (entrada rougher)
train_data['input_total'] = (
    train_data['rougher.input.feed_ag'] +
    train_data['rougher.input.feed_pb'] +
    train_data['rougher.input.feed_au'] +
    train_data['rougher.input.feed_sol']
)

# Estad√≠stica descriptiva
print(train_data['input_total'].describe())

# Detectar valores an√≥malos (suma == 0 o suma > 100)
anomalies = train_data[(train_data['input_total'] == 0) | (train_data['input_total'] > 100)]
print(f"Filas an√≥malas encontradas: {len(anomalies)}")
print(anomalies.head())

# Visualizaci√≥n de la distribuci√≥n
plt.figure(figsize=(10,5))
sns.histplot(train_data['input_total'], bins=50, kde=True)
plt.axvline(0, color='red', linestyle='--', label='Valor 0')
plt.axvline(100, color='orange', linestyle='--', label='Valor 100')
plt.title('Distribuci√≥n de la concentraci√≥n total en materia prima (input_total)')
plt.xlabel('Concentraci√≥n total (%)')
plt.legend()
plt.show()

# Eliminar filas an√≥malas
train_cleaned = train_data[~((train_data['input_total'] == 0) | (train_data['input_total'] > 100))].copy()
print(f"Filas despu√©s de limpieza: {len(train_cleaned)}")
```
